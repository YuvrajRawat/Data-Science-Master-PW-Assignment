```python
Question: What are the three measures of central tendency?
Answer: The three measures of central tendency are mean, median, and mode.

Question: What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a 
    dataset?
Answer: The mean is the average value of a dataset, the median is the middle value when the dataset is arranged in ascending or 
    descending order, and the mode is the most frequently occurring value. They provide a single value that represents the 
    center or typical value of the dataset.

Question: Measure the three measures of central tendency for the given height data: [178,177,176,177,178.2,178,175,179,180,175,
                                                                                     178.9,176.2,177,172.5,178,176.5]
Answer:

Mean: 177.8125
Median: 177.5
Mode: 178
    
Question: Find the standard deviation for the given data: [178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,
        176.5]
Answer: Approximately 1.689 (rounded to three decimal places).

Question: How are measures of dispersion such as range, variance, and standard deviation used to describe the spread of a 
    dataset? Provide an example.
Answer: Measures of dispersion provide information about how spread out the values in a dataset are from the central tendency 
    measures. For example, a larger range, variance, or standard deviation indicates greater variability in the data.

Question: What is a Venn diagram?
Answer: A Venn diagram is a graphical representation of the relationships between different sets of data, typically consisting 
    of overlapping circles or other shapes.

Question: For the two given sets A = {2,3,4,5,6,7} & B = {0,2,6,8,10}, find:
(i) A ∩ B
(ii) A ⋃ B
Answer:
(i) A ∩ B (intersection): {2, 6}
(ii) A ⋃ B (union): {0, 2, 3, 4, 5, 6, 7, 8, 10}

Question: What do you understand about skewness in data?
Answer: Skewness in data refers to the asymmetry in the distribution of data points.

Question: If a data is right skewed then what will be the position of median with respect to mean?
Answer: If a data is right skewed, the median will be less than the mean.

Question: Explain the difference between covariance and correlation. How are these measures used in statistical analysis?
Answer: Covariance measures the extent to which two variables change together, while correlation measures the strength and 
    direction of the linear relationship between two variables. They are used to analyze the relationship between variables in 
    statistical analysis.

Question: What is the formula for calculating the sample mean? Provide an example calculation for a dataset.
Answer: The formula for calculating the sample mean is: Mean = (Sum of all values) / (Number of values). For example, if the 
        dataset is {2, 4, 6, 8, 10}, then the mean = (2 + 4 + 6 + 8 + 10) / 5 = 6.

Question: For a normal distribution data what is the relationship between its measure of central tendency?
Answer: For a normal distribution, the mean, median, and mode are all equal and located at the center of the distribution.

Question: How is covariance different from correlation?
Answer: Covariance measures the direction of the linear relationship between two variables along with the extent to which they 
    change together, whereas correlation is a standardized measure representing the strength and direction of the linear 
    relationship between two variables.

Question: How do outliers affect measures of central tendency and dispersion? Provide an example.
Answer: Outliers can significantly affect measures of central tendency, such as the mean, pulling it towards the direction of 
    the outlier. They can also impact measures of dispersion, such as the range, variance, and standard deviation, making them 
    larger or smaller depending on the position of the outlier relative to the other data points. For example, in a dataset of
    exam scores where most students scored between 60 and 80, but one student scored 95, the mean would be pulled towards 95, 
    and the standard deviation might increase due to the increased spread caused by the outlier.
```
