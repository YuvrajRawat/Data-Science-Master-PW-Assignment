```python
Q1: Probability Mass Function (PMF) and Probability Density Function (PDF) are both used to describe the probabilities 
    associated with different values in a probability distribution.

Probability Mass Function (PMF): It applies to discrete random variables and gives the probability that a discrete random 
    variable is equal to a certain value. For example, consider a fair six-sided die. The PMF of rolling a 1 would be 1/6, as 
    there is one favorable outcome out of six possible outcomes.

Probability Density Function (PDF): It applies to continuous random variables and gives the probability density at a particular
    value. Unlike the PMF, the PDF doesn't directly give probabilities but rather probability densities. An example is the 
    normal distribution, where the PDF describes the shape of the bell curve and gives the density of probabilities at 
    different points along the curve.

Q2: Cumulative Density Function (CDF) represents the probability that a random variable takes on a value less than or equal to 
    a given value. It gives the cumulative probability up to a certain point.

For example, consider a standard six-sided die. The CDF for rolling a 3 would be the probability of rolling a 1, a 2, or a 3,
which is 3/6 = 0.5.

The CDF is used to understand the distribution of a random variable, its properties, and to calculate probabilities for ranges
of values.

Q3: The normal distribution is used as a model in various situations, including:

Height distribution of a population
Test scores
Errors in measurements
IQ scores
The parameters of the normal distribution, the mean (μ) and standard deviation (σ), determine the center and spread of the 
distribution, respectively. A larger standard deviation results in a wider distribution, while a larger mean shifts the 
\distribution to the right along the x-axis.

Q4: The normal distribution is important because:

It is a fundamental concept in statistics and probability theory.
It provides a theoretical framework for analyzing data and making statistical inferences.
Many real-life phenomena approximate a normal distribution, making it applicable in various fields.
It allows for the calculation of probabilities and confidence intervals.
Real-life examples of normal distribution include:

Heights of individuals in a population
IQ scores
Test scores
Blood pressure measurements
Q5: Bernoulli Distribution is a discrete probability distribution representing the outcomes of a single trial where there are
    only two possible outcomes: success or failure, typically denoted as 1 or 0.

An example of Bernoulli Distribution is flipping a fair coin, where success (1) might represent getting heads and failure (0)
getting tails.

Difference between Bernoulli Distribution and Binomial Distribution:

Bernoulli Distribution represents a single trial, while Binomial Distribution represents the number of successes in a fixed 
number of independent Bernoulli trials.


Q7: Uniform Distribution is a probability distribution where all outcomes are equally likely over a given range.

An example of uniform distribution is rolling a fair six-sided die, where each outcome (1, 2, 3, 4, 5, 6) has an equal 
probability of 1/6.

Q9: Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution as 
    the sample size increases, regardless of the shape of the population distribution.

The significance of the Central Limit Theorem lies in its application to inferential statistics. It allows us to make 
conclusions about a population based on a sample, even when the population distribution is not normal.

Q10: Assumptions of the Central Limit Theorem include:

The sample size should be sufficiently large (usually n ≥ 30).
The observations in the sample should be independent of each other.
The population should have a finite standard deviation.
```
