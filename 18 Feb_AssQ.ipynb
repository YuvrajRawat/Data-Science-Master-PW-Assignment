```python
Q1. What is an API? Give an example, where an API is used in real life.

An API (Application Programming Interface) is a set of rules and protocols that allows different software applications to 
communicate with each other. It defines how different components of software should interact, providing a way for developers to 
access the functionality of another software component or service without needing to understand its internal implementation.

An example of where an API is used in real life is the Google Maps API. Developers can use the Google Maps API to integrate 
maps and location-based services into their own applications. This allows developers to display maps, obtain directions, and 
perform geolocation functions within their own applications without having to build these functionalities from scratch.

Q2. Give advantages and disadvantages of using API.

Advantages of using API:

Reusability: APIs allow developers to reuse existing functionality and services, reducing development time and effort.
Scalability: APIs enable applications to scale more easily by allowing them to access external services and resources.
Flexibility: APIs provide a standardized way for applications to communicate, making it easier to integrate with third-party 
    services and adapt to changing requirements.
Innovation: APIs foster innovation by enabling developers to build on top of existing platforms and services, creating new and 
    innovative applications.
Disadvantages of using API:

Dependency: Applications built on top of APIs are dependent on the stability and availability of those APIs. Changes or downtime
    in the API can affect the functionality of dependent applications.
Security risks: APIs can expose sensitive data and functionality, increasing the risk of security vulnerabilities if not 
    properly secured and authenticated.
Complexity: Integrating with APIs and managing dependencies can add complexity to software development and maintenance.
Versioning issues: Changes to APIs may require updates to dependent applications, leading to compatibility and versioning issue
    s over time.
Q3. What is a Web API? Differentiate between API and Web API.

A Web API is an API that is accessed over the web using HTTP protocols. It provides a way for web services to expose their 
functionality to other applications and services over the internet. Web APIs typically follow RESTful principles and use 
standard HTTP methods like GET, POST, PUT, and DELETE for communication.

The main difference between API and Web API is the medium of communication. While both APIs and Web APIs define how software
components should interact, APIs can be accessed locally within the same system or network, while Web APIs are accessed over the
internet using standard web protocols like HTTP.

Q4. Explain REST and SOAP Architecture. Mention shortcomings of SOAP.

REST (Representational State Transfer): REST is an architectural style for designing networked applications. It relies on 
    stateless, client-server communication over standard protocols like HTTP. RESTful APIs use standard HTTP methods (GET, 
    POST, PUT,DELETE) to perform CRUD (Create, Read, Update, Delete) operations on resources. REST APIs typically use JSON or 
    XML as data interchange formats.

SOAP (Simple Object Access Protocol): SOAP is a protocol for exchanging structured information in the implementation of web 
    services. It relies on XML-based messaging formats and operates over various network protocols like HTTP, SMTP, and TCP. 
    SOAP APIs define a strict messaging structure and support features like security, transactions, and reliable messaging.

Shortcomings of SOAP:

Complexity: SOAP APIs tend to be more complex and verbose compared to RESTful APIs, making them harder to implement and 
    maintain.
Performance overhead: SOAP messages are often larger in size due to the XML format and additional metadata, leading to increased
    bandwidth and processing overhead.
Tight coupling: SOAP APIs can lead to tight coupling between client and server implementations, making it harder to evolve and
    scale the system.
Limited browser support: SOAP APIs are not as well-supported in web browsers compared to RESTful APIs, limiting their use in
    web-based applications.
Q5. Differentiate between REST and SOAP.

REST (Representational State Transfer):

Uses standard HTTP methods like GET, POST, PUT, DELETE for communication.
Supports multiple data interchange formats like JSON, XML.
Stateless communication.
Lightweight and simpler to implement.
Well-suited for web-based applications and mobile apps.
Loosely coupled architecture.
SOAP (Simple Object Access Protocol):

Operates over various network protocols like HTTP, SMTP, TCP.
Relies on XML-based messaging format.
Supports features like security, transactions, reliable messaging.
Heavier and more complex compared to REST.
Well-suited for enterprise applications and scenarios requiring strict standards compliance.
Tightly coupled architecture.
```
