```python
#Ans 1 Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a popular data visualization library for Python that is built on top of Matplotlib. 
Here are five common plots that can be created using Seaborn:

Scatter Plot:
A scatter plot is a two-dimensional plot that shows the relationship between two variables. 
It is used to identify the correlation between two variables. Seaborn's scatter plot is similar to Matplotlib's scatter 
plot but has additional features such as color-coding based on a third variable.

Line Plot: A line plot is used to visualize the trend of a continuous variable over time. 
Seaborn's line plot can display multiple lines on a single graph, making it useful for visualizing 
trends across multiple categories.

Bar Plot:
A bar plot is used to compare the frequency or distribution of categorical data. Seaborn's 
bar plot can display the mean, standard deviation, and confidence interval of each category, making it useful for 
visualizing statistical summaries.

Histogram:
A histogram is used to visualize the distribution of a continuous variable. Seaborn's 
histogram can display the probability density function (PDF) or the cumulative distribution function (CDF), making it 
useful for visualizing the distribution of data.

Heatmap:
A heatmap is used to visualize the correlation between variables in a matrix format. Seaborn's 
heatmap can display the correlation matrix using a color-coded scale, making it useful for visualizing patterns in large
datasets.
```

```python
#Ans 2 Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and 
#      y = "signal" for different events and regions.

import seaborn as sns
fmri_data = sns.load_dataset("fmri")
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")
```

```python
#Ans 3 Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', 
#      y = 'age' and y = 'fare'.

import seaborn as sns
titanic_data = sns.load_dataset("titanic")
sns.boxplot(data=titanic_data, x='pclass', y='age')
sns.boxplot(data=titanic_data, x='pclass', y='fare')
```

```python
#Ans 4 Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for 
#      the 'cut' column of the diamonds dataset.

import seaborn as sns
diamonds_data = sns.load_dataset("diamonds")
sns.histplot(data=diamonds_data, x='price', hue='cut')
```

```python
#Ans 5 Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

import seaborn as sns
iris_data = sns.load_dataset("iris")
sns.pairplot(data=iris_data, hue='species')
```

```python
#Ans 6 Use the "flights" dataset from seaborn to plot a heatmap.

import seaborn as sns
flights_data = sns.load_dataset("flights")
flights_data = flights_data.pivot('month', 'year', 'passengers')
sns.heatmap(data=flights_data, cmap='YlOrRd')
```
